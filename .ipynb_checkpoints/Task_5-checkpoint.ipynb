# Neural network - 2 hidden layers 

## Variables, parameters and hyper parameters

* z - stands for the linearity
* a - stands for activations - relu adn softma in the output layer
* w - weigths (parameters)
* b - bias
* beta1, beta2 - exponetial decay for the first and second time-steps
* lamda (L2)- streghten term for reularisation
* mom (v)-  coeficient of friction used in stochatic gradient descendent optimizer, SGD-mo
* minibatch - sizeof minibact for optimization
* dropout - proportion of neurons to drop -here I used inverted dropout all computation ar based in the keep probability (keep-prob)
* lr - learning rate
* v - friction coefficient(momemtum)/ decay average past(Adam)- same initiliazed to zero matrix are use - stimate of the first moment
* s  - estimation of the second gradient moment
* t -  time step used in Adam


## Functions

* Relu (activation) - squeezed linearity (z) to get the activation. Alpha (constant) 0.01 used to prevent nan cost
* Softmax  -  used to compute prediction probabilities 
* Cross entropy loss with log sofmax for computation of muticlass loss
* Truncated normal distribution - used to initilized weights and bias

## Optimizers and regularization

* Regularizers - L2 and droput
* optimizers SGD, SGD-Momentum, Adam 


## Weights initialisation - Initialisation

Weights can be initialized fron a truncated normal or from a normal distribution

# Class neural netwok 

The neural network can run with Adama optimizer, L2 regularisation, Adam, SGD with L2, SGD-Momentum +L2,
and droput, 

```python
def truncated_normal(mean=0, sd=1, low=0, upp=10):
        """weights value distribution"""
        return truncnorm((low - mean) / sd, (upp - mean) / sd, loc=mean, scale=sd)
    
```

##Â Neural network ANN2L - two layesr

```python
class ANN2L(object):
    """ Feedforward neural network / Multilayer perceptron classifier.
    """
    def __init__(self, hidden_n, l2=0., epochs=100, lr=0.001, lr_d = 0.90, schedule = 20, shuffle=True, 
                 minibatch_size=1, weights_init = "trunc_norm", dropout = False, drop_prob1=0, drop_prob2=0,
                 optimizer = "sgd", mom= 0.9, seed=None,):

        self.random = np.random.RandomState(seed)
        self.hidden_n = hidden_n
        self.l2 = l2
        self.epochs = epochs
        self.lr = lr
        self.lr_d = lr_d
        self.schedule = schedule
        self.shuffle = shuffle
        self.minibatch_size = minibatch_size
        self.weights_init = weights_init
        self.dropout = dropout
        self.drop_prob1 = drop_prob1
        self.drop_prob2 = drop_prob2
        self.optimizer = optimizer
        self.mom = mom
        
        if self.optimizer != "sgd_mo" and self.optimizer != "sgd" and self.optimizer != "adam":
            sys.exit("No an optimizer choice, select from 'sgd_mo', 'sgd' or 'adam'")
            
        if self.dropout == True and self.l2 != 0:
            sys.exit("Select either dropout or l2 regularisation")
            
        if self.weights_init != "trunc_norm" and self.weights_init != "rand_norm":
            sys.exit("No an initilisation choice, select from either 'trunc_nor' or 'rand_norm'")
                
        if (self.optimizer == "sgd"and self.mom != 0) or (self.optimizer == "adam"and self.mom != 0):
            sys.exit("Zero mom or change the optimizer to sgd_mo")
        

    def _onehot(self, y, n_classes):
        """Encode labels into one-hot representation
        """
        onehot = np.zeros((n_classes, y.shape[0]))
        for idx, val in enumerate(y.astype(int)):
            onehot[val, idx] = 1.
        return onehot.T
 
    def _relu(self, z):
        """Implements leakyrelu using a negative slope alpha""" 
        
        return np.maximum(z, 0.01*z)
        #  return np.maximum(z, np.zeros_like(z))
         
    def _relu_d(self, z):
        """Computes relu derivative with alpha to prevent nan"""  
        
        return np.where(z <= 0, z*0.01, 1)
  
    def _softmax(self,z):
        exp = np.exp(z )
        partition = np.sum(np.exp(z), axis=1, keepdims = True )
        
        return exp/partition

    def in_dropout(self, a, drop_prob):
        """Computes inverted dropout, 1-dropout
        keeps_prob is the maintained probability."""
        assert 0 <= drop_prob <= 1
        keep_prob = 1 -  drop_prob
        mask = np.random.uniform(0, 1.0, a.shape) < keep_prob
        
        if keep_prob > 0.0:
            scale = (1/keep_prob)
            return mask * a * scale
        #All elements kept
        if keep_prob == 1:
            return a
        #All elements dropped
        else:       
            return np.zeros_like(a)

    def _forward(self, X, drop_prob1, drop_prob2):
        """Compute forward propagation step"""

        # step net input to hidden layer1
     
        z_h1 = np.dot(X, self.w_h1) + self.b_h1
  
        #Optional dropout
        if self.dropout== True:
            a_h1 = self._relu(z_h1)
            a_h1 = self.in_dropout(a_h1, self.drop_prob1)
        else:
            a_h1 = self._relu(z_h1)

        # step hidden1 to hidden2
        z_h2 = np.dot(a_h1, self.w_h2) + self.b_h2 
         
         #Optional dropout
        if self.dropout== True:
            a_h2 = self._relu(z_h2)
            a_h2 = self.in_dropout(a_h2, self.drop_prob2)   
        else:
            a_h2 = self._relu(z_h2)

         # step hidden2 to output
        z_out = np.dot(a_h2, self.w_out) + self.b_out
        a_out = self._softmax(z_out)

        return z_h1, a_h1, z_h2, a_h2, z_out, a_out

    def _compute_cost(self,y_enc, output):
        """ If no regularisation the L2 term becomes zero"""
        L2_reg = self.l2*(np.sum(self.w_h1 ** 2.) +np.sum(self.w_h2 ** 2.) + 
                               np.sum(self.w_out ** 2.)) 
        ind = np.argmax(y_enc, axis = 1).astype(int)
        pred_probability = output[np.arange(len(output)), ind]
        log_preds = np.log(pred_probability)
        loss = -1.0 * np.sum(log_preds) / len(log_preds)
        loss+= L2_reg
        return loss

    def predict(self, X):
        """Predict class labels, X is the train, test or val dataset
        """
        z_h1, a_h1, z_h2, a_h2, z_out, a_out = self._forward(X, self.drop_prob1, self.drop_prob2)
        y_pred = np.argmax(a_out, axis=1)
        return y_pred

    def predict_val_test(self, X):
        """Predict class labels, X is the train, test or val dataset
        no dropout aplied to the test and validation dataset - probabilities 0 hard coded
        """
        z_h1, a_h1, z_h2, a_h2, z_out, a_out = self._forward(X, drop_prob1=0, drop_prob2=0)

        y_pred = np.argmax(a_out, axis=1)
        return y_pred
    
    def conf_matrix(self, y, y_pred):
        """Computes the confusion matrix cm target vs prediction"""      
        dim= len(np.unique(y))
        cm = np.zeros((dim, dim),int) 
        for i in range (len(y_pred)):
            target = y[i]
            pred = y_pred[i]
            cm[target, pred] += 1
        cm=(np.array(cm))
        return cm  
            
    def precision(self, y, conf_matrix):
        col = conf_matrix[:, y]
        return conf_matrix[y, y] / col.sum()
      
    def recall(self, y, conf_matrix):
        row = conf_matrix[y, :]
        return conf_matrix[y, y] / row.sum()
        
        
    def f1_score(self, precision, recall):
        F1 = 2 * (precision * recall) / (precision + recall)
        return F1
        

    def run(self, X_train, y_train, X_valid, y_valid, X_test, y_test):
        """ Learn weights from training data."""
        n_output = np.unique(y_train).shape[0]  # number of class labels     
        n_features = X_train.shape[1]
      
        if self.weights_init != "trunc_norm" and self.weights_init != "rand_norm":
            print("No an initilisation choice")  
            return self
         
        ########################
        # Weight initialization
        ########################
         
        if self.weights_init == "trunc_norm":
            print(" Weights initialised within a truncated normal distribution")
            # weights for input -> hidden
            rad_1= 1/np.sqrt(n_features)
            self.distw1 = truncated_normal(mean=0, sd=1, low=-rad_1, upp= rad_1)
            self.b_h1 = self.distw1.rvs((1,self.hidden_n[0]))
            self.w_h1 = self.distw1.rvs((n_features, self.hidden_n[0]))

            # weights for hidden -> hidden 2

            rad_2= 1/np.sqrt(self.hidden_n[0])
            self.distw2 = truncated_normal(mean=0, sd=1, low=-rad_2, upp=rad_2)
            self.b_h2 = self.distw2.rvs((1,self.hidden_n[1]))
            self.w_h2 = self.distw2.rvs((self.hidden_n[0], self.hidden_n[1]))

            # weights for hidden2 -> output

            rad_3= 1/np.sqrt(self.hidden_n[1])
            self.distw3 = truncated_normal(mean=0, sd=1, low=-rad_2, upp=rad_2)
            self.b_out = self.distw3.rvs((1,n_output))
            self.w_out = self.distw3.rvs((self.hidden_n[1], n_output))

        if self.weights_init == "rand_norm":
           
            print("Weights initialised within a normal distribution - 'He'")
            
            #  Input ---> first hidden layer
            rad_1= np.sqrt(2/(n_features))        
            self.w_h1 = self.random.normal(loc=0.0, scale= rad_1,size=(n_features, self.hidden_n[0]))
            self.b_h1 = np.zeros(self.hidden_n[0])
        
            #  first hidden --->second hidden layer
            
            rad_2= np.sqrt(2/self.hidden_n[0])
            self.w_h2 = self.random.normal(loc=0.0, scale= rad_2, size=(self.hidden_n[0], self.hidden_n[1]))
            self.b_h2 = np.zeros(self.hidden_n[1])
            
            #  2nd hidden -> output

            rad_3= np.sqrt(2/self.hidden_n[1])
            self.w_out = self.random.normal(loc=0.0, scale= rad_3, size=(self.hidden_n[1], n_output))
            self.b_out = np.zeros(n_output)
              
 
        ################################################
        # Initial velocities - momentum & adam optimizers
        ################################################

        self.vb_h1 = np.zeros_like(self.b_h1)        
        self.vw_h1 = np.zeros_like(self.w_h1)
        self.vb_h2 = np.zeros_like(self.b_h2)
        self.vw_h2 = np.zeros_like(self.w_h2)
        self.vb_out = np.zeros_like(self.b_out)
        self.vw_out = np.zeros_like(self.w_out)
        
        
        #Initializatin of sq for Adam

        self.sb_h1 = np.zeros_like(self.b_h1)   
        self.sw_h1 = np.zeros_like(self.w_h1)
        self.sb_h2 = np.zeros_like(self.b_h2)
        self.sw_h2 = np.zeros_like(self.w_h2)
        self.sb_out = np.zeros_like(self.b_out)
        self.sw_out = np.zeros_like(self.w_out)

        epoch_strlen = len(str(self.epochs))  # for progress formatting
        
        
        self.eval_ = {'cost': [], 'train_acc': [], 'valid_acc': [], 'test_acc':[]}

        y_train_enc = self._onehot(y_train, n_output) 

        # iterate over training epochs
 
        t= 0 # time -step for adam optimizer
       
        for i in range(self.epochs):
            if  i == self.schedule:
                for i in range(self.schedule, self.epochs, self.schedule):
                    self.lr = self.lr*self.lr_d
            
                    if self.lr <= 0.00001:
                        break
            
            # iterate over minibatches
            indices = np.arange(X_train.shape[0])

            if self.shuffle:
                self.random.shuffle(indices)

            for start_idx in range(0, indices.shape[0] - self.minibatch_size +
                                   1, self.minibatch_size):
                batch_idx = indices[start_idx:start_idx + self.minibatch_size]

                # forward propagation
                z_h1, a_h1, z_h2, a_h2, z_out, a_out = self._forward(X_train[batch_idx],
                                                        self.drop_prob1, self.drop_prob2)

                ##################
                # Backpropagation -SGD
                ##################

                # [n_samples, n_classlabels]
                error_out = (a_out - y_train_enc[batch_idx])/ a_out.shape[0]
                # error_out = (a_out - y_train_enc[batch_idx])

                # [n_samples, hidden_n]

                relu_derivative_h2= self._relu_d(a_h2)
                relu_derivative_h1= self._relu_d(a_h1)

                # [n_samples, n_classlabels] dot [n_classlabels, hidden-n]
                # -> [n_samples, hidden_n]
                error_h2 = (np.dot(error_out, self.w_out.T) *relu_derivative_h2)
                error_h1 = (np.dot(error_h2, self.w_h2.T) *relu_derivative_h1)

                
                # -> [hidden_n, n_classlabels]
                grad_w_out = np.dot(a_h2.T, error_out)
                grad_b_out = np.sum(error_out, axis=0)

                # [n_features, n_samples] dot [n_samples, hidden_n]
                # -> [n_features, hidden_n]
                grad_w_h2 = np.dot(a_h1.T, error_h2)
                grad_b_h2 = np.sum(error_h2, axis=0)

                # [n_features, n_samples] dot [n_samples, hidden_n]
                # -> [n_features, hidden_n]
                grad_w_h1 = np.dot(X_train[batch_idx].T, error_h1)
                grad_b_h1 = np.sum(error_h1, axis=0)

                # Regularization and weight updates- if not regularized  delta = grad)

                delta_w_out = (grad_w_out + self.l2*self.w_out*2)
                delta_b_out = grad_b_out  # bias is not regularized
               
                delta_w_h2 = (grad_w_h2 + self.l2*self.w_h2*2) #
                delta_b_h2 = grad_b_h2 # bias is not regularized

                delta_w_h1 = (grad_w_h1 + self.l2*self.w_h1*2)
                delta_b_h1 = grad_b_h1  # bias is not regularized
                
              
                #############
                # Update weights(parameters)
                #############
                t+=1  # Increase time-step before Adam
                
                if self.optimizer == "sgd_mo":
                    
                    self.vb_h1 = self.mom*self.vb_h1 - self.lr * delta_b_h1
                    self.vw_h1 = self.mom*self.vw_h1 - self.lr * delta_w_h1

                    self.vb_h2 = self.mom*self.vb_h2 - self.lr * delta_b_h2
                    self.vw_h2 = self.mom*self.vw_h2 - self.lr * delta_w_h2

                    self.vb_out = self.mom*self.vb_out - self.lr * delta_b_out
                    self.vw_out = self.mom*self.vw_out - self.lr * delta_w_out
                    
                    self.w_h1 += self.vw_h1
                    self.b_h1 += self.vb_h1
                    self.w_h2 += self.vw_h2
                    self.b_h2 += self.vb_h2
                    self.w_out += self.vw_out
                    self.b_out += self.vb_out

                if self.optimizer == "sgd":
                    
                    self.w_h1 -= self.lr * delta_w_h1
                    self.b_h1 -= self.lr * delta_b_h1

                    self.w_h2 -= self.lr * delta_w_h2
                    self.b_h2 -= self.lr * delta_b_h2
                      
                    self.w_out -= self.lr * delta_w_out
                    self.b_out -= self.lr * delta_b_out
                    
                if self.optimizer == "adam":
                    """b1: decay for previous gradient, 
                    b2:decay for the prevous squared gradient"""
                    
                    eps_sta = 1e-8
                    beta1 = 0.9
                    beta2 = 0.999
                    
                    ##############
                    # average of the gradients and updates layer h1
                    #############
                    # weights
       
                    self.vw_h1 = beta1 * self.vw_h1  + (1 - beta1) * delta_w_h1
                    self.sw_h1 = beta2 * self.sw_h1 + (1.- beta2) * np.square(delta_w_h1)
                    self.vw_h1_c = self.vw_h1 / (1 - beta1** t)
                    self.sw_h1_c = self.sw_h1 / (1. - beta2 ** t)
 
                    #bias
                    self.vb_h1 = beta1 * self.vb_h1  + (1 - beta1) * delta_b_h1
                    self.sb_h1 = beta2 * self.sb_h1 + (1.- beta2) * np.square(delta_b_h1)
            
                    self.vb_h1_c = self.vb_h1 / (1 - beta1** t)
                    self.sb_h1_c = self.sb_h1 / (1. - beta2 ** t)
                    
                    # params update
       
                    prop_wh1 = self.lr * self.vw_h1_c / (np.sqrt(self.sw_h1_c) + eps_sta)
                    prop_bh1 = self.lr * self.vb_h1_c / (np.sqrt(self.sb_h1_c) + eps_sta)
   
                    self.w_h1-= prop_wh1
                    
                    self.b_h1-= prop_bh1
                    
                    #############
                    # layer h2
                    ###########
                    
                    #weigths
                    self.vw_h2 = beta1 * self.vw_h2  + (1 - beta1) * delta_w_h2
                    self.sw_h2 = beta2 * self.sw_h2 + (1.- beta2) * np.square(delta_w_h2)
                    self.vw_h2_c = self.vw_h2 / (1 - beta1** t)
                    self.sw_h2_c = self.sw_h2 / (1. - beta2 ** t)
                    
                    # bias
                    
                    self.vb_h2 = beta1 * self.vb_h2  + (1 - beta1) * delta_b_h2
                    self.sb_h2 = beta2 * self.sb_h2 + (1.- beta2) * np.square(delta_b_h2)
                    self.vb_h2_c = self.vb_h2 / (1 - beta1** t)
                    self.sb_h2_c = self.sb_h2 / (1. - beta2 ** t)
                    
 
                    # params update
                    prop_wh2 = self.lr * self.vw_h2_c / (np.sqrt(self.sw_h2_c) + eps_sta)
                    prop_bh2 = self.lr * self.vb_h2_c / (np.sqrt(self.sb_h2_c) + eps_sta)

                    self.w_h2-= prop_wh2
                    self.b_h2-= prop_bh2
        
        
                    #############
                    # output layer
                    ###########

                    
                    #weigths
                    self.vw_out = beta1 * self.vw_out  + (1 - beta1) * delta_w_out
                    self.sw_out = beta2 * self.sw_out + (1.- beta2) * np.square(delta_w_out)
                    
                    self.vw_out_c = self.vw_out / (1 - beta1** t)
                    self.sw_out_c = self.sw_out / (1. - beta2 ** t)
                    
                    
                    # layer bias
                    
                    self.vb_out = beta1 * self.vb_out  + (1 - beta1) * delta_b_out
                    self.sb_out = beta2 * self.sb_out + (1.- beta2) * np.square(delta_b_out)
                    
                    self.vb_out_c = self.vb_out/ (1 - beta1** t)
                    self.sb_out_c = self.sb_out / (1. - beta2 ** t)
                    
                     # params update
                    prop_w_out = self.lr * self.vw_out_c / (np.sqrt(self.sw_out_c) + eps_sta)
                    prop_b_out = self.lr * self.vb_out_c / (np.sqrt(self.sb_out_c) + eps_sta)

                    self.w_out-= prop_w_out
                    self.b_out-= prop_b_out

            #############
            # Evaluation
            #############

            # Evaluation after each epoch during training
            z_h1, a_h1, z_h2, a_h2, z_out, a_out = self._forward(X_train, self.drop_prob1,
                                                                 self.drop_prob2)
  
            cost = self._compute_cost(y_enc=y_train_enc, output = a_out)
        
            if np.isnan(cost) == True:
                break

            y_train_pred = self.predict(X_train)
            if self.dropout:
                y_valid_pred = self.predict_val_test(X_valid)
                y_test_pred = self.predict_val_test(X_test)
            else:
                y_valid_pred = self.predict(X_valid)
                y_test_pred = self.predict(X_test)

            train_acc = ((np.sum(y_train == y_train_pred)).astype(np.float) /
                            X_train.shape[0])
            valid_acc = ((np.sum(y_valid == y_valid_pred)).astype(np.float) /
                            X_valid.shape[0])
            test_acc = (np.sum(y_test == y_test_pred).astype(np.float) / X_test.shape[0])
            
            if test_acc <= 0.80* train_acc:
                
                print("Early stopping")
                return self

            sys.stderr.write('\r%0*d/%d | Train cost: %.2f ''| Train/Valid Acc.: %.2f%%/%.2f%% '
                              '|Test Acc.: %.2f%%''| L rate.: %.5f '%
                            (epoch_strlen, i+1, self.epochs, cost, train_acc*100,
                             valid_acc*100, test_acc*100, self.lr))
            sys.stderr.flush()

            self.eval_['cost'].append(cost)
            self.eval_['train_acc'].append(train_acc)
            self.eval_['valid_acc'].append(valid_acc)
            self.eval_['test_acc'].append(test_acc)
        
        return self
```

```python
total_sample =0
for i in range(8):
    row_sum = np.sum(cm[i:i+1,:])
    
    total_sample = total_sample + row_sum
    print('label: %.0f '' |  Precision: %.2f%%'' |  Recall: %.2f%%'' |  f1-score: %.2f%%'' |  support: %.0f' %
        (i, ann.precision(i, cm), ann.recall(i, cm), ann.f1_score(ann.precision(i, cm),ann.recall(i, cm)),row_sum))
print("Total samples:", total_sample)
```
